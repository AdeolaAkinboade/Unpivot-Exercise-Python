```python
#I used the data and decided to make a Pandas DataFrame from a dictionary to reaarange everything
```

```python
the_data = {'Row Labels': ['08 05 2023', '09 05 2023', 'Grand Total'],
            'FOSMIDDELMALL':['6','8.1','14.1'],
            'FOSPTANORTH':['15.5','10.7','26.2'],
            'FOSROSEBANK':['12','9.2','21.2'],
            'Grand Total': ['33.5','28','61.5']
            }
```

```python
df_1= pd.DataFrame(the_data) #dataframe for the dictionary created
```

```python
df_1
```

```python
#unpivoting the dataframe using melt 
df_unpivoted= pd.melt(df_1, id_vars= 'Row Labels',var_name='Machine Name' ,value_vars=['FOSMIDDELMALL','FOSPTANORTH','FOSROSEBANK','Grand Total'],value_name='Hours Off')
```

```python
df_unpivoted
```

```python
#dropped rows with 'Grand Total'
df_unpivoted.drop(df_unpivoted.index[8:12], inplace= True)
```

```python
df_unpivoted
```

```python
#Dropped the Grand Total Rows
df_unpivoted.drop(df_unpivoted.index[2],inplace= True)
```

```python
df_unpivoted.drop(df_unpivoted.index[4],inplace= True)
```

```python
df_unpivoted
```

```python
#reaaranged the Row Labels column so that the dates were in ascending order
df_unpivoted.sort_values(by= 'Row Labels',ascending=True)
```

```python
#renamed Row Label Column to the Date Column
df_unpivoted.rename(columns= {'Row Labels':'Date'}, inplace= True)
```
